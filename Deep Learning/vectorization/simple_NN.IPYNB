```python
import numpy as np

input_size = 3  # Number of input features
hidden_layer1_size = 5  # Number of neurons in the first hidden layer
hidden_layer2_size = 4  # Number of neurons in the second hidden layer
hidden_layer3_size = 3  # Number of neurons in the third hidden layer
output_size = 2  # Number of output neurons


def initialize_parameters():
    np.random.seed(0)  # Set a seed for reproducibility
    
    # Initialize weights and biases for each layer
    W1 = np.random.randn(hidden_layer1_size, input_size) * 0.01
    b1 = np.zeros((hidden_layer1_size, 1))
    
    W2 = np.random.randn(hidden_layer2_size, hidden_layer1_size) * 0.01
    b2 = np.zeros((hidden_layer2_size, 1))
    
    W3 = np.random.randn(hidden_layer3_size, hidden_layer2_size) * 0.01
    b3 = np.zeros((hidden_layer3_size, 1))
    
    W4 = np.random.randn(output_size, hidden_layer3_size) * 0.01
    b4 = np.zeros((output_size, 1))
    
    parameters = {
        "W1": W1,
        "b1": b1,
        "W2": W2,
        "b2": b2,
        "W3": W3,
        "b3": b3,
        "W4": W4,
        "b4": b4
    }
    
    return parameters
def tanh_activation(Z):
    return np.tanh(Z)

def forward_propagation(X, parameters):
    # Retrieve the parameters from the dictionary
    W1, b1, W2, b2, W3, b3, W4, b4 = parameters
    
    # Forward pass through each layer
    Z1 = np.dot(W1, X) + b1
    A1 = tanh_activation(Z1)
    
    Z2 = np.dot(W2, A1) + b2
    A2 = tanh_activation(Z2)
    
    Z3 = np.dot(W3, A2) + b3
    A3 = tanh_activation(Z3)
    
    Z4 = np.dot(W4, A3) + b4
    A4 = np.exp(Z4) / np.sum(np.exp(Z4), axis=0)  # Softmax activation for the output
    
    cache = {
        "Z1": Z1,
        "A1": A1,
        "Z2": Z2,
        "A2": A2,
        "Z3": Z3,
        "A3": A3,
        "Z4": Z4,
        "A4": A4
    }
    
    return A4, cache


# Sample input data (replace with your data)
X = np.random.randn(input_size, 1).astype(np.float64)  # Ensure data type is float64

# Initialize the parameters
parameters = initialize_parameters()

# Perform forward propagation
output, cache = forward_propagation(X, parameters)

# 'output' contains the final output of the network



```
