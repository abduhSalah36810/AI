```python
import matplotlib.pyplot as plt

# Sample data after mean centering
data = [
    [8, 7.4, 2],
    [-7, -7.6, -3],
    [0, 0.4, 0],
    [-14, -12.6, -6],
    [13, 12.4, 7]
]

# Principal components
PC1 = [0.6, 0.7, 0.3]
PC2 = [0.3, -0.6, 0.7]

# Projecting data onto the 2D space defined by PC1 and PC2
projected_data = []
for point in data:
    proj_point = [point[0] * PC1[0] + point[1] * PC1[1] + point[2] * PC1[2],
                  point[0] * PC2[0] + point[1] * PC2[1] + point[2] * PC2[2]]
    projected_data.append(proj_point)

# Plotting the data points in the 2D space
plt.figure(figsize=(8, 6))
plt.scatter([point[0] for point in projected_data],
            [point[1] for point in projected_data], c='b', label='Data Points')

plt.xlabel('Principal Component 1 (PC1)')
plt.ylabel('Principal Component 2 (PC2)')
plt.title('Data Points in 2D Space After PCA')
plt.legend()
plt.grid()
plt.show()
```
