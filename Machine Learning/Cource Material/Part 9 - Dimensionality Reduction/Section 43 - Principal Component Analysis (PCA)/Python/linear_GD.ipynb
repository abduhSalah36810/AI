```python
import numpy as np
import matplotlib.pyplot as plt

# Sample dataset
x = np.array([1, 2, 3, 4])
y = np.array([2, 3, 4, 5])

# Initial parameters
theta0 = 0.5
theta1 = 0.5

# Hypothesis function
hypothesis = lambda x: theta0 + theta1 * x

# Learning rate
alpha = 0.1

# Number of iterations
num_iterations = 2

# Gradient descent
for _ in range(num_iterations):
    for i in range(len(x)):
        error = hypothesis(x[i]) - y[i]
        theta0 -= alpha * (1/len(x)) * error
        theta1 -= alpha * (1/len(x)) * error * x[i]

# Plotting the data points and the linear regression line
plt.scatter(x, y, color='blue', label='Data Points')
plt.plot(x, hypothesis(x), color='red', label='Linear Regression')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Regression with Gradient Descent')
plt.legend()

plt.show()
```
